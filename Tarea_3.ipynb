```python
import cv2
import os
import numpy as np
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image as ins_get_image
from sklearn.neighbors import KNeighborsClassifier, kneighbors_graph
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python


app = FaceAnalysis(providers=['CUDAExecutionProvider', 'CPUExecutionProvider'])
app.prepare(ctx_id=0, det_size=(640, 640))
images = []
embeddings = []
labels = []
# using readlines()

with open("/content/drive/MyDrive/Tarea_3/T3 Dataset/data.txt") as fp:
    Lines = fp.readlines()
    for line in Lines:
        images.append(line.split("\t")[0])
        labels.append(line.split("\t")[1].replace("\n", ""))
print(labels)
i = 0
for image in images:
  i += 1
  img = cv2.imread(image)
  # print(img)
  faces = app.g/*et(img)

  for face in faces:
    embeddings.append(face["embedding"])




```

```python
k = 3  # Puedes ajustar este valor (número de vecinos)
knn = KNeighborsClassifier(n_neighbors=k)

# Entrenar el modelo con tus embeddings y etiquetas
X = embeddings  # embeddings generados por InsightFace
Y = labels  # etiquetas de identidad de las caras
print(labels)
knn.fit(X, Y)
```

```python
app = FaceAnalysis(providers=['CUDAExecutionProvider', 'CPUExecutionProvider'])
app.prepare(ctx_id=0, det_size=(640, 640))
images_test = []
embeddings_test = []
labels_test = []
# using readlines()

with open("/content/drive/MyDrive/Tarea_3/T3 Dataset/Test/data.txt") as fp:
    Lines = fp.readlines()
    for line in Lines:
        images_test.append(line.split("\t")[0])
        labels_test.append(line.split("\t")[1].replace("\n", ""))
print(labels_test)
i = 0
for image in images_test:
  i += 1
  img = cv2.imread(image)
  # print(img)
  faces = app.get(img)
  if(len(faces) > 1):
    print(i)
    # print(len(faces))

  for face in faces:
    embeddings_test.append(face["embedding"])
```

```python
X_test = embeddings_test  # embeddings generados por InsightFace
Y_test = labels_test  # etiquetas de identidad de las caras
print(Y_test)
# print(len(X_test), len(Y_test))
knn.score(X_test, Y_test)
```

```python
# kneighbors_graph(X_test, 3)
total = 0
num = 0
prediccion = knn.predict(X_test)
for i in range(len(prediccion)):
  num +=1
  if(prediccion[i] == Y_test[i]):
    total += 1
print(total)
print(num)
print(total/num)
for i in range(len(Y_test)):
  print(prediccion[i], Y_test[i])
```

```python
#Opcion 1
confusion_matrix_a = confusion_matrix(Y_test, prediccion)

# cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

# cm_display.plot()
# plt.show()

#Opcion 2
unified_labels = ['Francisco De-la-maza', 'Gonzalo Vicente', 'Jaime Fernandez', 'Jorge Plaza', 'Jose Queirolo', 'Manuel Bentjerodt', 'Maria Marin', 'Raimundo Martinez', 'Rodrigo Quezada', 'Sebastian Diaz', 'Sebastian Ma-shichoy', 'Vicente Gana']
fig, ax = plt.subplots(figsize=(7, 7))
ax.matshow(confusion_matrix_a)
for i in range(confusion_matrix_a.shape[0]):
  for j in range(confusion_matrix_a.shape[1]):
    ax.text(x=j, y=i,s=round(confusion_matrix_a[i, j], 2), va='center', ha='center', size='xx-small')
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Simple Model Confusion Matrix', fontsize=18)
plt.xticks(range(12), labels=unified_labels, rotation=90)
plt.yticks(range(12), labels=unified_labels)
#plt.savefig("matrix_de_confusion_simple.png")
plt.show()
```

```python
total_images = np.sum(confusion_matrix_a, axis = 1, keepdims = True)

# rmc_resnet = confusion_matrix_a.astype(np.float32) / np.sum(confusion_matrix_a, axis = 1, keepdims = True)




accuracy_per_class = []
for i in range(len(confusion_matrix_a[0])):
  if(total_images[i] == 0):
    accuracy_per_class.append(0)
    continue
  accuracy_per_class.append(confusion_matrix_a[i][i]/total_images[i])
print(accuracy_per_class)
print(sum(accuracy_per_class)/len(accuracy_per_class))

print(accuracy_per_class)

fig, ax = plt.subplots(figsize = (6,6))
ax.bar(unified_labels, accuracy_per_class)
plt.xticks(range(12), labels=unified_labels, rotation=90)
#plt.yticks(range(12), labels=y)
for i, v in enumerate(accuracy_per_class):
    ax.text(i, v+0.01, str(np.round(v,3))[1:-1], ha='center')
plt.show()
```
